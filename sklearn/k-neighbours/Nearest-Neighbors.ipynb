```python
import numpy as np
import matplotlib.pyplot as plot

%matplotlib notebook
```

```python
# Definimos vector con numpy de puntos de 2 dimensiones
X_train = np.array([[1,1], [2,2.5], [3,1.2], [5.5,6.3], [6,9], [7,6], [7,7], [7,8]])

#etiquetamos cada punto definido
# Definimos array con la categoría /etiqueta a la que pertenece cada punto
Y_train = ['red', 'red', 'red', 'blue', 'blue', 'blue','blue','blue']
```

```python

# Definimos nueva figura de matplotlib
plot.figure()

# Pintamos los puntos en las coordenadas definidas
plot.scatter(X_train[:,0], X_train[:,1], s = 170, color = Y_train[:])

plot.show()
```

```python

#creamos un punto de prueba en las coordenadas [3,4]
#el objetivo luego va a ser clasificar este punto,si va con los azules o con los rojos

X_test = np.array([3,4])
plot.figure()
plot.scatter(X_train[:,0], X_train[:,1], s = 170, color = Y_train[:])
plot.scatter(X_test[0], X_test[1], s = 170, color = 'green')
plot.show()

```

```python
#Para ejecutar el clasificador basado en el vecino más cercano primero hay que medir la distancia entre 2 puntos
#podemos usar la distancia euclidea

def distancia_euclidea(x, y):
    return np.sqrt(np.sum((x - y)**2)) # np.sqrt and np.sum son funciones de numpy para trabajar con arrays
```

```python
#Para cada punto tenemos que medir la distancia con el punto a testear
#Para cada punto de los datos de entrenamiento,calculamos la distancia al punto de prueba
num = len(X_train) # Obtenemos el número de puntos de entrenamiento
distancias = np.zeros(num) # Inicializar vector de distancias a zeros
for i in range(num):
    distancias[i] = distancia_euclidea(X_train[i], X_test) # obtener distancia euclidea entre el punto de entrenamiento y el de test
print(distancias)
```

```python
#del vector de distancias,obtenemos la distancia mínima y su etiqueta correspondiente
min_index = np.argmin(distancias) # Obtener el índice con la distancia más pequeña
print(Y_train[min_index])
```

```python
#creamos un punto de prueba en las coordenadas [4,5]
X_test = np.array([4,5])
plot.figure()
plot.scatter(X_train[:,0], X_train[:,1], s = 170, color = Y_train[:])
plot.scatter(X_test[0], X_test[1], s = 170, color = 'orange')
plot.show()
```

```python
#Para cada punto tenemos que medir la distancia con el punto a testear
#Para cada punto de los datos de entrenamiento,calculamos la distancia al punto de prueba
num = len(X_train) # Obtenemos el número de puntos de entrenamiento
distancias = np.zeros(num) # Inicializar vector de distancias a zeros
for i in range(num):
    distancias[i] = distancia_euclidea(X_train[i], X_test) # obtener distancia euclidea entre el punto de entrenamiento y el de test
print(distancias)
```

```python
#Now we can find the minimum distance, and the label for that point
min_index = np.argmin(distancias) # Get the index with smallest distance
print(Y_train[min_index])
```
